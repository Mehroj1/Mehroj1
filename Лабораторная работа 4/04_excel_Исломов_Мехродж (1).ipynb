2. Случайным образом выберите 5% строк из каждой таблицы и сохраните две таблицы на разные листы в один файл `recipes.xlsx`. Дайте листам названия "Рецепты" и "Отзывы", соответствующие содержанию таблиц. 

```python
import pandas as pd
import numpy as np

reviews = pd.read_csv('reviews_sample.csv', index_col=0)
recipes = pd.read_csv('recipes_sample.csv', index_col=0)

# выбор 5% случайных строк из каждой таблицы
reviews_sample = reviews.sample(frac=0.05)
recipes_sample = recipes.sample(frac=0.05)

with pd.ExcelWriter('recipes.xlsx') as writer:
    reviews_sample.to_excel(writer, sheet_name='Отзывы')
    recipes_sample.to_excel(writer, sheet_name='Рецепты')
```

3. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_assign`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи присваивания массива значений диапазону ячеек.

```python
import xlwings as xw
import pandas as pd

recipes_df = pd.read_csv("recipes_sample.csv", index_col=0, usecols=["id", "name", "minutes", "submitted", "description", "n_ingredients"])

recipes_df["seconds_assign"] = recipes_df["minutes"] * 60

# открываем
wb = xw.Book("recipes.xlsx")
sheet = wb.sheets["Рецепты"]

sheet.range("G1").value = "seconds_assign"
sheet.range("G2").options(transpose=True).value = recipes_df["seconds_assign"].values

wb.save()
wb.close()
```

```python
df = pd.read_excel('recipes.xlsx', sheet_name=1)
df
```

4. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_formula`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи формул Excel.

```python
import xlwings as xw

wb = xw.Book("recipes.xlsx")
sheet = wb.sheets["Рецепты"]

sheet.range("I1").value = "seconds_formula"
sheet.range("I2:I1500").formula = "=C2*60"

wb.save()
wb.close()
```

5. Сделайте названия всех добавленных столбцов полужирными и выровняйте по центру ячейки.

```python
import xlwings as xw

# Открываем файл
wb = xw.Book("recipes.xlsx")
sheet = wb.sheets["Рецепты"]

sheet.range("G1").api.Font.Bold = True
sheet.range("G1").api.HorizontalAlignment = xw.constants.HAlign.xlHAlignCenter

sheet.range("I1").api.Font.Bold = True
sheet.range("I1").api.HorizontalAlignment = xw.constants.HAlign.xlHAlignCenter

# Сохраняем и закрываем файл
wb.save()
wb.close()
```

6. Раскрасьте ячейки столбца `minutes` в соответствии со следующим правилом: если рецепт выполняется быстрее 5 минут, то цвет - зеленый; от 5 до 10 минут - жёлтый; и больше 10 - красный.

```python
import time
import xlwings as xw

# Открываем файл
wb = xw.Book("recipes.xlsx")
sheet = wb.sheets["Рецепты"]

# Получаем диапазон столбца minutes
minutes = sheet.range("C2:C100")

# Цикл по каждой ячейке столбца minutes
for cell in minutes:
    # Получаем значение ячейки
    value = cell.value
    
    if value is None:  # если значение ячейки пустое, устанавливаем значение 0
        value = 0
        
    # Преобразуем значение ячейки в число
    value = float(value)
    
    # Раскрашиваем ячейку в зависимости от значения
    if value < 5:
        cell.colorindex = (0, 255, 0) # Зеленый
    elif value < 10:
        cell.colorindex = (255, 255, 0)# Желтый
    else:
        cell.colorindex = (255,0,0) # Красный
    
    # Добавляем задержку в 0.1 секунду
    time.sleep(0.1)

# Сохраняем и закрываем файл
wb.save()
wb.close()
```

7. Добавьте на лист `Рецепты`  столбец `n_reviews`, содержащий кол-во отзывов для этого рецепта. Выполните задание при помощи формул Excel.

```python
import xlwings as xw

# открываем файл с таблицами
wb = xw.Book('recipes.xlsx')
sht_recipes = wb.sheets['Рецепты']

# добавляем столбец
sht_recipes.range('J1').value = 'n_reviews'
sht_recipes.range('J2:J100').formula = '=COUNTIF(OFFSET(Reviews!$B$2,0,0,COUNTA(Reviews!$B:$B)-1,1),B2)'

# сохраняем изменения и закрываем файл
wb.save()
wb.close()
```
