Считайте, что в слове есть опечатка, если данное слово не содержится в списке `words`. 

```python
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        raw_data = file.read()
        result = chardet.detect(raw_data)
        return result['encoding']

file_path = 'litw-win.txt'
encoding = detect_encoding(file_path)
print(encoding)
```

```python
with open('litw-win.txt', 'r', encoding=encoding) as file:
    words = file.read().split()
import Levenshtein

def correct_typo(word, words):
    if word not in words:
        min_distance = float('inf')
        closest_word = ''
        for w in words:
            distance = Levenshtein.distance(word, w)
            if distance < min_distance:
                min_distance = distance
                closest_word = w
        return closest_word
    else:
        return word

sentence = "Thiss is a sentennce with typose."
corrected_sentence = ' '.join([correct_typo(word, words) for word in sentence.split()])
print(corrected_sentence)
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize

# Загрузка данных из файла CSV
data = pd.read_csv("recipes_sample.csv")

# Создание пустого списка для уникальных слов
words = []

# Итерация по описаниям рецептов
for description in data['description']:
    if isinstance(description, str):  # Проверка типа данных
        # Токенизация текста описания
        tokens = word_tokenize(description)
        # Добавление уникальных слов в список words
        words.extend([word.lower() for word in tokens if word.isalpha()])

# Получение набора уникальных слов
unique_words = set(words)

# Вывод результатов
print("Уникальные слова:", unique_words)
```

1.3 Напишите функцию, которая для заданного слова `word` возвращает `k` ближайших к нему слов из списка `words` (близость слов измеряется с помощью расстояния Левенштейна)

```python
import Levenshtein

def find_k_nearest_words(word, words, k):
    # Вычисление расстояния Левенштейна для каждого слова из списка words
    distances = [(w, Levenshtein.distance(word, w)) for w in words]

    # Сортировка слов по расстоянию Левенштейна
    sorted_words = sorted(distances, key=lambda x: x[1])

    # Получение k ближайших слов
    nearest_words = [w[0] for w in sorted_words[:k]]

    return nearest_words
```

```python
# Список слов
words = unique_words

# Заданное слово
word = "puffed"

# Количество ближайших слов, которые нужно найти
k = 3

# Поиск k ближайших слов к заданному слову
nearest_words = find_k_nearest_words(word, words, k)

# Вывод результатов
print(f"Ближайшие {k} слова к '{word}': {nearest_words}")
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords

# Загрузка стоп-слов
nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))

# Инициализация стеммера
stemmer = SnowballStemmer('russian')

# Предполагая, что у вас уже есть данные рецептов, загруженные в переменную 'data'

# Создание пустых списков для слов, стемм и нормализованных слов
words = []
stemmed_words = []
normalized_words = []

# Итерация по описаниям рецептов
for description in data['description']:
    # Проверка, что описание является строкой
    if isinstance(description, str):
        # Токенизация текста описания
        tokens = word_tokenize(description, language='russian')
        
        # Удаление стоп-слов и добавление слов в список
        filtered_words = [word for word in tokens if word.isalpha() and word.lower() not in stop_words]
        words.extend(filtered_words)
        
        # Стемминг и добавление стеммированных слов в список
        stemmed_words.extend([stemmer.stem(word) for word in filtered_words])
        
        # Нормализация и добавление нормализованных слов в список
        normalized_words.extend([word.lower() for word in filtered_words])

# Создание DataFrame на основе списков слов, стемм и нормализации
df = pd.DataFrame({'слово': words, 'стеммированное_слово': stemmed_words, 'нормализованное_слово': normalized_words})

# Расчет доли стоп-слов в общем количестве слов
total_words = len(words)
stop_words_count = sum(word.lower() in stop_words for word in words)
stop_words_ratio = stop_words_count / total_words

# Вывод доли стоп-слов
print(f"Доля стоп-слов: {stop_words_ratio:.2%}")

# Расчет и вывод топ-10 самых часто употребляемых слов до удаления стоп-слов
top_words_before = df['слово'].value_counts().head(10)
print("Топ-10 слов до удаления стоп-слов:")
print(top_words_before)

# Удаление стоп-слов из списка слов
df_filtered = df[~df['слово'].isin(stop_words)]

# Расчет и вывод топ-10 самых часто употребляемых слов после удаления стоп-слов
top_words_after = df_filtered['слово'].value_counts().head(10)
print("\nТоп-10 слов после удаления стоп-слов:")
print(top_words_after)
```

3.2 Вычислите близость между каждой парой рецептов, выбранных в задании 3.1, используя косинусное расстояние (`scipy.spatial.distance.cosine`) Результаты оформите в виде таблицы `pd.DataFrame`. В качестве названий строк и столбцов используйте названия рецептов.

```python
import pandas as pd
from scipy.spatial.distance import cosine

# Предполагая, что у вас уже есть список случайных рецептов random_recipes

# Создание пустой таблицы для результатов
similarity_matrix = pd.DataFrame(index=random_recipes, columns=random_recipes)

# Вычисление близости между парами рецептов
for i in range(len(random_recipes)):
    for j in range(i, len(random_recipes)):
        # Вычисление косинусного расстояния между векторами
        similarity = 1 - cosine(vectors[i].toarray(), vectors[j].toarray())
        # Запись значения близости в таблицу
        similarity_matrix.iloc[i, j] = similarity
        similarity_matrix.iloc[j, i] = similarity

# Вывод таблицы с результатами
print(similarity_matrix)
```

3.3 Какие рецепты являются наиболее похожими? Прокомментируйте результат (словами).

```python
import numpy as np


similarity_array = similarity_matrix.to_numpy()

most_similar_indices = np.argmax(similarity_array, axis=1)

recipe_names = similarity_matrix.index

result_df = pd.DataFrame(index=recipe_names, columns=['Most Similar Recipe', 'Similarity Score'])

for i, recipe in enumerate(recipe_names):
    most_similar_recipe = recipe_names[most_similar_indices[i]]
    similarity_score = similarity_array[i, most_similar_indices[i]]
    result_df.loc[recipe] = [most_similar_recipe, similarity_score]

print(result_df)
```
